```python
columns = [
    'No.', '상품소분류', '상 품 명', 'OFF 수량', 'ON 수량', 'TOTAL 수량', '기간 개월수', '월판매량 수량',
    '현재고 수량', '발주분 수량', '총재고 수량', '재고월수', '3개월후 재고', '발주희망', '발주수량 병',
    '발주수량 박스단위', '예상발주 박스', '발주확정 박스', '발주확정 병'
]

# 빈 데이터프레임 생성
df = pd.DataFrame(columns=columns)
```

```python
df['상 품 명'] = off_array
df['OFF 수량'] = df_off['합 계'].values[1:-2]
```

```python
## on 수량 dictionalry로 만들기
on_dict = {}
for i in range(len(df_on)):
    if df_on['품목명'][i] in on_array:
        on_dict[df_on['품목명'][i]] = df_on['EA'][i]
        
remain_dict = {}
for i in range(len(df_remain)):
    if df_remain[df_remain.columns[1]][i] in remain_array:
        #column1 = 품목, column 9 = 수량
        remain_dict[df_remain[df_remain.columns[1]][i]] = df_remain[df_remain.columns[9]][i]
```

```python
df['ON 수량'] = [on_dict[result_on_off[name]] if name in result_on_off else 0 for name in off_array]
df['현재고 수량'] = [remain_dict[result_off_remain[name]] if name in result_off_remain else 0 for name in off_array]
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from openpyxl.utils import get_column_letter
# 2. pandas로 엑셀 파일로 저장
output_file = 'result_with_formulas_1.xlsx'
df.to_excel(output_file, index=False, engine='openpyxl')

# 3. openpyxl로 엑셀 파일 불러오기
wb = load_workbook(output_file)
ws = wb.active


# 5. 엑셀 파일 저장
wb.save(output_file)

# 노란색 채우기 스타일 정의 (3개월 후 재고가 음수인 경우)
yellow_fill = PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')

# 12개월, 월 판매량, 총재고, 재고월수, 3개월 후 재고 등 수식 적용
for row in range(2, ws.max_row + 1):
    # TOTAL 수량 계산: OFF 수량 + ON 수량
    ws[f'F{row}'] = f"=D{row}+E{row}"
    
    # 기간 개월수는 12로 고정
    ws[f'G{row}'] = 12
    
    # 월 판매량 계산: TOTAL 수량 / 기간 개월수 (소수점 버림)
    ws[f'H{row}'] = f"=TRUNC(F{row}/G{row})"
    
    # 총재고는 현재고 수량과 동일
    ws[f'K{row}'] = f"=I{row}"
    
    # 재고월수 계산: 현재고 수량 / 월 판매량
    ws[f'L{row}'] = f"=ROUND(I{row}/H{row},2)"
    
    # 3개월 후 재고 계산: 총재고 - 3 * 월 판매량
    ws[f'M{row}'] = f"=K{row}-3*H{row}"
    
    # 발주희망 계산: 월 판매량 * 12 - 3개월 후 재고
    ws[f'N{row}'] = f"=H{row}*12 - M{row}"
    
    # 발주수량 병: 발주희망과 동일
    ws[f'O{row}'] = f"=N{row}"
    
    # 박스 단위: 12로 고정
    ws[f'P{row}'] = 12
    
    # 예상발주 박스: 발주수량 병 / 박스 단위
    ws[f'Q{row}'] = f"=O{row}/P{row}"
    
    # 발주확정 박스: 예상발주 박스와 동일
    ws[f'R{row}'] = f"=Q{row}"
    
    # 발주확정 병: 발주확정 박스 * 12
    ws[f'S{row}'] = f"=R{row}*12"
    
    # # 3개월 후 재고가 음수일 경우 노란색으로 표시
    # if ws[f'M{row}'].value is not None and int(ws[f'M{row}'].value) < 0:
    #     ws[f'M{row}'].fill = yellow_fill

# 마지막 행에 각 열의 합계를 추가
sum_row = ws.max_row + 1
for col in range(4, ws.max_column + 1):  # D열부터 S열까지 합계 계산
    col_letter = get_column_letter(col)
    ws[f'{col_letter}{sum_row}'] = f"=SUM({col_letter}2:{col_letter}{ws.max_row-1})"

# 컬럼 너비 자동 조정
for column_cells in ws.columns:
    max_length = 0
    column = column_cells[0].column_letter  # 해당 컬럼 이름(A, B, C...)
    for cell in column_cells:
        try:
            if cell.value:
                max_length = max(max_length, len(str(cell.value)))
        except:
            pass
    adjusted_width = (max_length + 2)
    ws.column_dimensions[column].width = adjusted_width

# 엑셀 파일 저장
wb.save(output_file)

print(f"{output_file} 파일이 생성되었고, 엑셀 함수 및 서식이 적용되었습니다.")
```
